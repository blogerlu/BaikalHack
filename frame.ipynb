```python
from Bio import SeqIO


def search_frame(path):
    record = SeqIO.read('D:/BaikalHack/Scripts/Data/new_data/' + path, "fasta")
    table = 11
    min_pro_len = 70
    ff = ''

    for strand, nuc in [(+1, record.seq), (-1, record.seq.reverse_complement())]:
        for frame in range(3):
            length = 3 * ((len(record)-frame) // 3) #Multiple of three
            for pro in nuc[frame:frame+length].translate(table).split("*"):
                if len(pro) >= min_pro_len:
                    print("%s...%s - length %i, strand %i, frame %i" \
                          % (pro[:30], pro[-3:], len(pro), strand, frame))
            if len(pro) > len(ff):
                ff = pro
    return ff
```

```python
import os
```

```python
dirf = os.listdir('D:/BaikalHack/Scripts/Data/new_data/')
```

```python
for d in dirf:
    s = search_frame(d)
    file = open('D:/BaikalHack/Scripts/Data/frames/' + d + '_frame.fasta', 'w')
    file.writelines('>' + d[:-6])
    file.writelines('\n')
    file.writelines(s)
    file.close()
```
